```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url="https://www.failory.com/startups/india"
```

```python
page=requests.get(url)
page
```

```python
page.status_code
```

```python
htmlCode=page.text
```

```python
soup= BeautifulSoup(htmlCode)
soup
```

```python
abc=[]
n=soup.find_all("ul",{"role":"list"})
for i in n:
    abc.append(i.text)
string=str(abc)
```

```python
company=[]
ven=[]
ci=[]
b=soup.find_all("h3")
for jk in b:
    ven.append(jk.text)
c=str(ven)
c
incv=r"\) [A-Za-z0-9. [\(A-Za-z\-)&!]*"
compani=re.findall(incv,c)
compani
for i in compani:
    company.append(i)

print(company)

```

```python
len(company)
```

```python
cities=[]
ci=r"\s[A-Za-z]+Started"
city=re.findall(ci,string)
for i in city:
    cities.append(i.split('S')[0])
print(cities)
```

```python
len(cities)
```

```python
starts=[]
star=r"Started in: [0-9]+"
started=re.findall(star,string)
for i in started:
    x=(i.split(":")[1])
    #f=x.strip("'")
    starts.append(x)
print(starts)
```

```python
len(starts)
```

```python
founders=[]
k=[]
fou=r"\:\s[0-9]{4}Founders:\s([a-zA-Z\s[a-zA-Z,\.]*\:)"
founder=re.findall(fou,string)
for i in founder:
    founders.append(i.split("I")[0])
#for j in founders:
   # m=j.split(",")
   # k.append(len(m))
print(founders)
```

```python
len(founders)
```

```python
industries=[]
inc=r"Industries: +([a-z0-9A-Z\-[a-zA-Z]*)"
industry=re.findall(inc,string)
for i in industry:
    industries.append(i)
industries
```

```python
len(industries)
```

```python
no_of_employees=[]
em=r"Number of employees: +([0-9]*-[0-9]*)"
emplo=re.findall(em,string)
for i in emplo:
    v=i.strip("][")
    no_of_employees.append(v)
no_of_employees
```

```python
len(no_of_employees)
```

```python
funding_amount=[]
f=r"Funding amount:\s\$+([0-9]+,[0-9]+,[0-9]+)"
fun=re.findall(f,string)
for i in fun:
    #v=i.strip("][")
    #k=v.replace(",","")
    funding_amount.append(i)
funding_amount
```

```python
len(funding_amount)
```

```python
funding_rounds=[]
fr="Number of funding rounds: [0-9]+"
fnr=re.findall(fr,string)
for i in fnr:
    #cdf=i.split(':')[1]
    #billa=cdf.strip(" ' '")
    funding_rounds.append(i)
funding_rounds
```

```python
len(funding_rounds)
```

```python
number_of_investors=[]
inv=r"Number of investors: [0-9]+"
inve=re.findall(inv,string)
for i in inve:
    #har=i.split(":")[1]
    #ram=har.strip(" ' '")
    number_of_investors.append(i)
number_of_investors
```

```python
len(number_of_investors)
```
